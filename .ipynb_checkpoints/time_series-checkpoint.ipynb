### Create embeddings

```python
# Window size 
M =  5

df = data.iloc[:,1:4]

x = []
for i in range(0,df.shape[0]-M):
    x.append(np.array(df.iloc[i:i+M,:]).flatten())
x = np.vstack(x)
```

```python
plt.scatter(x[0,:],x[3,:])
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Scale the data before apaplying PCA

scaler = StandardScaler()

#create copy of DataFrame
scaled_df=x.copy()

#created scaled version of DataFrame
scaled_df=pd.DataFrame(scaler.fit_transform(scaled_df))

#define PCA model to use
pca = PCA(n_components=3)

#fit PCA model to data
pca_fit = pca.fit(scaled_df,)
pcs = pca.transform(scaled_df)
```

```python
# Plot PCA output for PC2, PC2,PC3

for i in range(6):
    fig = plt.figure(figsize=(20, 8))
    ax0 = fig.add_subplot(1, 2, 1, projection='3d')
    t = ax0.scatter(pcs[:,0], pcs[:,1], pcs[:,2], c = data.iloc[:pcs.shape[0], i + 1], cmap=plt.get_cmap("plasma"), s=5)
    plt.colorbar(t)

    plt.title(data.columns[i+1])
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    ax0.set_zlabel("PC3")
    #plt.savefig("images/PCA_"+data.columns[i+1])
```

### PCA VS TIME

```python
plt.plot(range(pcs[:,0].shape[0]),pcs[:,0],label="pca1")
plt.plot(range(pcs[:,0].shape[0]),pcs[:,1],label="pca2")
plt.plot(range(pcs[:,0].shape[0]),pcs[:,2],label="pca3")
plt.xlabel("Time")
plt.ylabel("PCAs")
plt.title("PCA VS Time")
plt.legend()
```
