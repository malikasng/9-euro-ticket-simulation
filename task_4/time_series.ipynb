```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as clt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from modules.nonlinear_approx import *
from modules.radial_basis import *
```

# Predict the area density for different ROIs in the subway car

```python
def get_utilization(file,area_list,M=5, show_arc=False,savePath='fig'):
    """ Get the utilization for different regions.

    Args:
        file (str): Input file with utilization data
        area_list (list): List of regions to consider
        M (int, optional): Length of window when creating the embedding space. Defaults to 5.
        show_arc (bool, optional): Displays the graph of velocity VS arc length of the passengers. Defaults to False.
        savePath (str, optional): Path to the output file. Defaults to 'fig'.
    """

    # Read data
    data = pd.read_csv(file,sep=' ',header=1)[:]

    # Create Embeddings space. This has been implemented similar to Task5  of Assignment 5.  
    df = data.iloc[:,area_list]
    x = []
    for i in range(0,df.shape[0]-M):
        x.append(np.array(df.iloc[i:i+M,:]).flatten())
    x = np.vstack(x)

    # Apply PCA
    # Scale the data before apaplying PCA
    scaler = StandardScaler()

    #create copy of DataFrame
    scaled_df=x.copy()

    #created scaled version of DataFrame
    scaled_df=pd.DataFrame(scaler.fit_transform(scaled_df))

    #define PCA model to use
    pca = PCA(n_components=len(area_list))

    #fit PCA model to data
    pca_fit = pca.fit(scaled_df,)
    pcs = pca.transform(scaled_df)


    # Non-linear approximation of Velocity using radial basis function.
    delta_t =1
    x_data = pcs[1:,0]
    x0 = pcs[-1:,0]
    v = pcs[1:,0]-pcs[:-1,0] / delta_t

    # epsilon =0.4
    # L =1000
    x1_est_t = []
    x_data = x_data.reshape(-1, 1)
    x_range, v_est, plot_func, epsilon = find_nonlinear_approx(x_data, v, 1000, 0.4)
    
    if show_arc:
        plt.plot(x_data, v_est)
        plt.title("Arc length velocity over arclength")
        plt.xlabel("Arc length")
        plt.ylabel("Arc length velocity")
        plt.show()
        plt.savefig(savePath+'_arclen.png')
        
    
    # Approximate the utilization of a region using Radial Basis function
    
    # L = 1000 , epsilon=0.4
    r_rbf = RBF(1000,0.4)
    r_rbf.fit(v,data.iloc[:len(v),area_list[0]])
    pred = r_rbf.predict(v_est)

    plt.plot(data.iloc[:,area_list[0]],lw=0.25)
    plt.title("Actual Utilization")
    plt.xlabel("Time")
    plt.ylabel("Utiization")
    plt.savefig(savePath+'_actual.png')
    plt.show()


    plt.plot(pred,lw=0.5)
    plt.title("Future Predicted Utilization")
    plt.xlabel("Time")
    plt.ylabel("Utiization")
    plt.savefig(savePath+'_predict.png')
    plt.show()
```
